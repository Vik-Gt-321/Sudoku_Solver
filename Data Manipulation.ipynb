```python
import numpy as np
```

```python
import cv2
```

```python
import os
```

```python
training_data = []
path = "C:/0-9/"

dirs = os.listdir(path)

for sub in dirs:
    
    num = int(sub[-2:]) -1 
    print(num)
    dig_type_folder = path + sub + "/"
#     print(dig_type_folder)
    
    img_dir = os.listdir(dig_type_folder)
    for img in img_dir:
#         print(img)
        img_path = dig_type_folder + img
        img = cv2.imread(img_path)
        img = np.array(img)
        training_data.append([img, num])
```
